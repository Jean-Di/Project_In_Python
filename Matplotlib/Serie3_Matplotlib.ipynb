### Exercise 5: Create a scatter plot showing the relationship between study hours and scores: Study Hours: [1, 2, 3, 4, 5, 6, 7, 8] and Scores: [50, 55, 60, 65, 70, 75, 80, 82]. Add axis labels and a title, and include a trend line.

```python
import numpy as np
Study_Hours =[1, 2, 3, 4, 5, 6, 7, 8]
Scores=[50, 55, 60, 65, 70, 75, 80, 82]
plt.scatter(Study_Hours,Scores)

a, b = np.polyfit(Study_Hours, Scores, 1)
plt.plot(Study_Hours, a * np.array(Study_Hours) + b, color='green')


plt.title('Study Hours vs. Scores')
plt.xlabel('Study Hours')
plt.ylabel('Scores')
plt.show()
```

# Problem Serie 3: Matplotlib and Pandas Library

### Set up of data and libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_excel('data_serie3.xlsx')
data.head(3)
```

#### Exercise 1: Analyze the share of total energy production and consumption over the years for countries. Create pie graph to visualize these share.

```python
plt.figure(figsize=(20, 8))
for country in data['Country']:
    plt.pie(data['Total Energy Production (TWh)'],labels=data['Country'], autopct='%1.1f%%')
plt.title('Share of Total Energy Production(2023) among countries')

plt.figure(figsize=(20, 8))
for country in data['Country']:
    plt.pie(data['Total Energy Consumption (TWh)'],labels=data['Country'], autopct='%1.1f%%')
plt.title('Share of Total Energy Consumption (2023) among countries')

plt.show()
```

#### Exercise 2: Rank the countries based on their energy efficiency and visualize the rankings in a horizontal bar chart. Highlight the top three and bottom three countries in terms of energy efficiency

```python
data.head(2)
```

```python
rank= data.sort_values(by='Energy Efficiency (%)', ascending=False)

plt.figure(figsize=(10, 6))
plt.barh(rank['Country'], rank['Energy Efficiency (%)'], color='skyblue')
plt.title('Energy Efficiency Rankings by Country (2023)')
plt.xlabel('Energy Efficiency (%)')
plt.axvline(x=rank['Energy Efficiency (%)'].max(), color='red', linestyle='--', label='Top Performer')
plt.axvline(x=rank['Energy Efficiency (%)'].min(), color='blue', linestyle='--', label='Bottom Performer')
plt.legend()
plt.grid(axis='x')
plt.show()
```

#### Exercise 3: Investigate the relationship between the percentage of renewable energy in total production and CO2 emissions. Create a scatter plot to illustrate this relationship and fit a regression line.

```python
# Scatter plot
plt.scatter(data['Renewable Energy (% of Total)'], data['CO2 Emissions (Mt)'], color='green')

# Fit regression line
a, b = np.polyfit(data['Renewable Energy (% of Total)'], data['CO2 Emissions (Mt)'], 1)
plt.plot(data['Renewable Energy (% of Total)'], a * data['Renewable Energy (% of Total)'] + b, color='red')

plt.title('Renewable Energy Percentage against CO2 Emissions')
plt.xlabel('Renewable Energy (%)')
plt.ylabel('CO2 Emissions (Mt)')
plt.grid()
plt.show()
```

#### Exercise 4: Analyze how population size affects total energy consumption. Create a scatter plot of population versus total energy consumption and calculate the correlation coefficient.

```python
plt.scatter(data['Population (millions)'],data['Total Energy Consumption (TWh)'], color='blue')
correlation = data['Population (millions)'].corr(data['Total Energy Consumption (TWh)'])
plt.title('Population on Total Energy Consumption')
plt.xlabel('Population (millions)')
plt.ylabel('Total Energy Consumption (TWh)')
plt.grid()
plt.show()
print('The correlation between them is : ',correlation)
```

#### Exercise5:  Assuming a 4% annual increase in energy consumption for the next 5 years, project the future energy consumption for each country. Present this data in a table format.

#### Exercise 6: Compare the energy systems of two selected countries by examining their energy production, consumption, efficiency, and CO2 emissions. Create a grouped bar chart for visual comparison.

```python
import random
choice1 = random.choice(data['Country'])
choice2 = random.choice(data['Country'])

countries = [choice1, choice2]
comparison_data = data[data['Country'].isin(countries)]

x = np.arange(len(countries)) 
fig, ax = plt.subplots(figsize=(10, 5))
bars1 = ax.bar(x,comparison_data['Total Energy Production (TWh)'], label='Production (TWh)')
bars2 = ax.bar(x,comparison_data['Total Energy Consumption (TWh)'], label='Consumption (TWh)')
bars3 = ax.bar(x,comparison_data['CO2 Emissions (Mt)'], label='CO2 Emissions (Mt)')

ax.set_ylabel('Values')
ax.set_title('Energy Comparison between countries')
ax.set_xticks(x)
ax.set_xticklabels(countries)
ax.legend()

plt.show()
```

#### Exercise 7: Calculate the total renewable energy produced by each country and visualize it in a pie chart to show the distribution of renewable energy across the countries.

```python
re=data['Total Energy Production (TWh)']*data['Renewable Energy (% of Total)']/100
data['Total Renewable Energy Produced']=re

plt.figure(figsize=(20,8))
plt.pie(data['Total Renewable Energy Produced'], labels=data['Country'], autopct='%1.1f%%')
plt.title('Distribution of Renewable Energy across the Countries')
plt.show()
```

#### Exercise 8: Examine the gap between total energy production and consumption for each country. Present this information using a horizontal bar chart.

```python
plt.figure(figsize=(8,8))
gap = data['Total Energy Production (TWh)']-data['Total Energy Consumption (TWh)']
plt.bar(data['Country'], gap)
plt.xticks(rotation=45)
plt.grid()
plt.title('Gap between Total Energy Production and Consumption')
plt.ylabel('Gap')
plt.xlabel('Country')
plt.show()
```

#### Exercise 9: Hypothetically assess how a 10% increase in renewable energy production would affect CO2 emissions for each country. Calculate and visualize the new CO2 emissions.

```python
increase_re= data['Renewable Energy (% of Total)']+ (0.1*data['Renewable Energy (% of Total)'])
new_co2= increase_re*data['CO2 Emissions (Mt)']/data['Renewable Energy (% of Total)']
data['New CO2 Emissions (Mt)']=new_co2

plt.figure(figsize=(8, 6))
plt.bar(data['Country'], data['New CO2 Emissions (Mt)'], color='skyblue')
plt.title('New CO2 Emissions After 10% Increase in Renewable Energy Production')
plt.xlabel('Countries')
plt.ylabel('New CO2 Emissions (Mt)')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

#### Exercise 10: Analyze the relationships among Total Energy Production, Population, and CO2 emissions for each country. Create three scatter plots to visualize these relationships: Energy Production vs. CO2 Emissions, Population vs. CO2 Emissions and Energy Production vs. Population.

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Energy Production vs. CO2 Emissions
axs[0].scatter(data['Total Energy Production (TWh)'], data['CO2 Emissions (Mt)'], color='blue')
axs[0].set_title('Energy Production  vs. CO2 Emissions')
axs[0].set_xlabel('Total Energy Production (TWh)')
axs[0].set_ylabel('CO2 Emissions (Mt)')
axs[0].grid()

# Population vs. CO2 Emissions
axs[1].scatter(data['Population (millions)'], data['CO2 Emissions (Mt)'], color='orange')
axs[1].set_title('Population vs. CO2 Emissions')
axs[1].set_xlabel('Population (millions)')
axs[1].set_ylabel('CO2 Emissions (Mt)')
axs[1].grid()

# Energy Production vs. Population
axs[2].scatter(data['Total Energy Production (TWh)'], data['Population (millions)'], color='green')
axs[2].set_title('Energy Production vs. Population')
axs[2].set_xlabel('Total Energy Production (TWh)')
axs[2].set_ylabel('Population (millions)')
axs[2].grid()
```
