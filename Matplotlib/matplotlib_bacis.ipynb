#### *Matplotlib* is a Python library for data visualization that allows you to create static, animated, and interactive plots in a variety of formats

##### Before you start plotting, you need to import the library. It is commonly imported as plt: 

import matplotlib.pyplot as plt

### Basic Concepts

#### 1. *Figure and Axes*
   - *Figure*: The overall window or area where the graph is drawn.
   - *Axes*: The coordinate system (X and Y axes) on which data is plotted.

#### 2. *Plotting a Simple Graph*
You can create a simple line plot using the plot() function.

```python
x = [1, 2, 3, 4, 5]
y = [2, 9, 5, 7, 11]

plt.plot(x, y)
plt.title("Y against X")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()  # Displays the plot
```

### Types of Graphs

#### 1. *Line Graph*
Used to show trends over time.

```python
plt.plot(x, y, marker='o')  # Adds markers
plt.show()
```

#### 2. *Bar Chart*
Used to compare values.

```python
categories = ['A', 'B', 'C']
values = [10, 20, 15]

plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

#### 3. *Histogram*
Used to show the distribution of a dataset.

```python
data = [1, 2, 2, 3, 3, 3, 4, 4, 5]
plt.hist(data, bins=5)
plt.title("Histogram")
plt.show()
```

#### 4. *Scatter Plot*
Used to show the relationship between two variables.

```python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

#### 5. *Pie Chart*
Used to show proportions of a whole.

```python
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes,labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")
plt.show()
```

### Creating Multiple Plots in One Figure

You can create multiple plots in a single figure using subplot():

```python
plt.subplot(2, 2, 1)  # 2 rows, 2 columns, 1st subplot
plt.plot(x, y)
plt.title("Plot 1")

plt.subplot(2, 2, 2)  # 2nd subplot
plt.bar(categories, values)
plt.title("Plot 2")

plt.subplot(2, 2, 3)  # 3rd subplot
plt.scatter(x, y)
plt.title("Plot 3")

plt.subplot(2, 2, 4)  # 4th subplot
plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Pie Chart")

plt.tight_layout()  # Adjusts spacing
plt.savefig('my_subplot.pdf')  # Save as PDF
plt.show()
```

### Saving Plots

You can save plots as various file formats, including PDF:

```python
plt.plot(x, y)
plt.title("Line Plot")
plt.savefig('my_plot.pdf')  # Save as PDF
plt.savefig('my_plot.png')   # Save as PNG
```

### Customizing Graphs

#### 1. *Colors and Styles*
 You can customize colors and line styles.

```python
plt.plot(x, y, color='red', linestyle='--', linewidth=2)
```

#### 2. *Adding Legends*
Use plt.legend() to add a legend to your plot

```python
plt.plot(x, y, label='Data 1')
plt.plot(x, [i**2 for i in x], label='Data 2')
plt.legend()
```

#### 3. *Annotations*
You can annotate points in your plot.

```python
plt.plot(x, y, label='Data 1')
plt.plot(x, [i**2 for i in x], label='Data 2')
plt.legend()

plt.annotate('Highest Point', xy=(4, 7), xytext=(3, 8),arrowprops=dict(facecolor='black', shrink=0.05))
```

### Complete Example

```python
import matplotlib.pyplot as plt

# Data
x = [5, 2, 3, 4, 6]
y = [7, 9, 5, 7, 11]

# Plot
plt.plot(x, y, marker='o', color='blue', linestyle='-', linewidth=2, label='Data Line')

# Customization
plt.title("Example Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

# Save the plot
plt.savefig('example_plot.png')

# Show the plot
plt.show()
```
