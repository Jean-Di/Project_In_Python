### 5. Rename the column names as you want

### 6. Extract the period between 8:00 AM and 5:00 PM (8h00 and 17h00)

```python
from datetime import time
start_hour=time(8,0,0)
end_hour=time(17,59,0)
data_extract.iloc[:, 1] = pd.to_datetime(data_extract.iloc[:, 1], format='%H:%M:%S').dt.time
donnees_filtrees = data_extract[(data_extract.iloc[:, 1] >= start_hour) & (data_extract.iloc[:, 1] <= end_hour)]

donnees_filtrees
```

### 7. Remove missing data

```python
# 1st Method : Direct replacement
data_cleaned= donnees_filtrees.dropna()
data_cleaned
```

```python
# 2nd Method : Passing by nan value before replacement
import numpy as np
donnees_filtrees = donnees_filtrees.replace('---', np.nan)  # Remplace '---' par des valeurs NaN

complet_data = donnees_filtrees.dropna()  # Supprime les lignes avec des valeurs NaN
complet_data 
```

### 8. Calculate the correlation between solar radiation and other variables

```python
#Setup the data for the correlation ==> remove datetime value for that
data_correlation=complet_data.iloc[:,[2,3,4,5,6,7,8]]
correlation = data_correlation.corr()['Solar']
print(correlation)
```

```python
corr = data_cleaned['Prcpt'].corr(data_cleaned['Solar'])
corr
```

##### Why the correlation between prcpt and solar is nan (not a number) ??
When you observe that the correlation between two variables is NaN (Not a Number), it mainly means that there is insufficient valid data to compute the correlation. Some other possible reasons for this are:

###### 1. Missing Data
If either of the two columns contains missing values (NaN), the correlation function cannot produce a valid result.
  
###### 2. Constant Values
If one of the two columns has constant values (here, all values in "Rain" are 0), there is no variance in that column, making the correlation undefined.

###### 3. Data Types
Ensure that the columns involved in the correlation calculation are numeric types. If they are of object or string type, the correlation cannot be computed.

###### *How to Check ?*

To diagnose the issue, you can perform the following checks:

1. *Check for Missing Values*:
   print(df[['...', '...']].isnull().sum())
   

2. *Check for Variance*:
   print(df[['...', '...']].var())
   

3. *Check Data Types*:
   print(df[['...', '...']].dtypes)

### 9. Keep only the variables with a correlation greater than 0.3

```python
correlated_vars = correlation[correlation.abs() > 0.3].index.tolist()
data_final = data_cleaned[correlated_vars]
data_final
```

### 10. Plot graphs of correlated variables (scatter plot) to visualize the relationships between solar radiation and the remaining variables.

```python
import matplotlib.pyplot as plt
for var in correlated_vars:
       plt.figure(figsize=(10, 6))
       plt.scatter(data_final['Solar'], data_final[var])
       plt.title(f'Solar Radiation vs. {var}')
       plt.xlabel('Solar Radiation')
       plt.ylabel(var)
       plt.grid()
       plt.show()
```
